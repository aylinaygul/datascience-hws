```python
from sklearn import datasets
#importing iris dataset
iris = datasets.load_iris()
```

```python
# Creating dataframe of imported iris dataset
import pandas as pd
data=pd.DataFrame({
    'sepal length':iris.data[:,0],
    'sepal width':iris.data[:,1],
    'petal length':iris.data[:,2],
    'petal width':iris.data[:,3],
    'species':iris.target
})
```

```python
#Q1
from sklearn.model_selection import train_test_split

X=data[['sepal length', 'sepal width', 'petal length', 'petal width']]  #Features
y=data['species']  #Labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1) 
#Splitting dataset as 75% training and 25% test and randomizing
```

```python
#Importing Random Forest Model
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=10) #initializing model
model.fit(X_train,y_train) #training model with training sets
```

```python
#Q2
#predicting the species of each flower in the test set
y_pred=model.predict(X_test)
```

```python
#Importing scikit-learn metrics module for accuracy calculation
from sklearn import metrics
#printing accuracy calculation result
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Importing scikit-learn metrics module for confusion matrix
from sklearn.metrics import confusion_matrix

confusion_mat = confusion_matrix(y_test, y_pred) #creating confusion matrix with test and predict values
print(confusion_mat)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
#classification report that includes precision, recall and f1 score
```

```python
#3
estimator = model.estimators_[2]
#second tree
```

```python
from sklearn.tree import export_graphviz #exporting graphviz library for visualization
```

```python
# Exporting second tree as dot file
export_graphviz(estimator, out_file='tree.dot', 
                feature_names = iris.feature_names,
                class_names = iris.target_names,
                rounded = True, proportion = False, 
                precision = 2, filled = True)
```

```python
#conversion to png from dot file
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=600'])
```
